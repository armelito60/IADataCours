```python
import pandas as pd
import yfinance as yf
from yahoofinancials import YahooFinancials
import pandas_datareader as web
import numpy as np
```

```python
tickers = ["AAPL","GOOGL","AMZN", "NKE"]
multpl_stocks = web.get_data_yahoo(tickers,
start = "2021-09-01",
end = "2021-10-01")
```

```python
multpl_stocks["Open"]
```

```python
portfolio_returns = multpl_stocks["Open"].pct_change()
```

```python
portfolio_returns.head(10)
```

```python
#On initialise des poids au hasard (qui seront optimisés par la suite)
weight_portfolio = [0.1, 0.5, 0.3, 0.7]
return_port = portfolio_returns.dot(weight_portfolio)
return_port.head(10)
```

```python
#Matrice de covariance
covariance_matrix = portfolio_returns.cov()*252
covariance_matrix
```

```python
# Calcul de la variance du portefeuille
portfolio_variance = np.transpose(weight_portfolio)@covariance_matrix@weight_portfolio
portfolio_variance
```

```python
# Calcul de la volatilité (risque)
portfolio_volatility = np.sqrt(portfolio_variance)
print("Portfolio Variance is", portfolio_variance)
print("Portfolio Volatility is", portfolio_volatility)
```
