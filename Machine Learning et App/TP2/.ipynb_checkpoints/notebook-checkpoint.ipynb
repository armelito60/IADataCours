```python
import pandas as pd
import numpy as np

df = pd.read_csv("SALARY.csv")
df.head()
```

```python
df = df.drop('chiffre', axis=1)
```

```python
df.head()
```

```python
#Avant d'effectuer une visualisation des données, on séparer notre base en appretissage/test
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

#On récupère nos X et nos y
X = df.iloc[:,1:13]
y = df.iloc[:,13:14]

#On sépare nos données en train et en test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# transformation des labels en quantitatives avec Label Encoder sur la base train

y_train = y_train.apply(LabelEncoder().fit_transform)
y_train
```

```python
from sklearn.preprocessing import OneHotEncoder

X_train = pd.concat([X_train, pd.get_dummies(X_train['genre'], prefix='genre')], axis=1)
X_train.drop(['genre'], axis=1, inplace=True)

X_train = pd.concat([X_train, pd.get_dummies(X_train['statut'], prefix='statut')], axis=1)
X_train.drop(['statut'], axis=1, inplace=True)

X_train = pd.concat([X_train, pd.get_dummies(X_train['etudes'], prefix='etudes')], axis=1)
X_train.drop(['etudes'], axis=1, inplace=True)

X_train = pd.concat([X_train, pd.get_dummies(X_train['genre'], prefix='genre')], axis=1)
X_train.drop(['genre'], axis=1, inplace=True)
```
