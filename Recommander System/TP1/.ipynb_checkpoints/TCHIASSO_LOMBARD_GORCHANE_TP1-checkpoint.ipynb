# Part 2 - Building a simple recommender

### 1. Choose a metric (or score) to rate the movies on.
Nous choisissons la métric de la plus haute note attribuée sur une base de 50 votants avec une durée comprise entre 45 et 300min.


```python
#Import des données dans le but d'afficher les colonnes qui nous intéressent
film_df = pd.read_csv("movies_metadata.csv")
film_df.dtypes
```

```python
#Import des données dans le but d'afficher les colonnes qui nous intéressent
film_df = pd.read_csv("movies_metadata.csv", usecols = ["vote_count", "vote_average", "runtime","title","popularity"])
film_df.dtypes
```

```python
film_df
```

```python
m= film_df['vote_count'].quantile(0.8)
print(m)
```

Dans notre base de données, 80% des films ont un nombre de vote au moins égale à 50, c'est pour cela que nous utilisons ce nombre en  critère de sélection.

### 2. Decide on the prerequisites for the movie to be featured on the chart.

Les films qui apparaîtrons sur le chart seront ceux dont le nombre de vote sera égal ou supérieur à 50 avec une durée comprise entre 45min et 300min

### 3. Calculate the score for every movie that satisfies the conditions.

```python
#On créer la base de données avec un nombre de vote supérieur à 50
df_mask=film_df['vote_count']>=50
condition_df = film_df[df_mask]
condition_df
```

```python
#On selectionne les films ayant une durée supérieur à 45 min et inférieure à 300min

df_mask=film_df['runtime']>=45
condition1_df = condition_df[df_mask]

df_mask=film_df['runtime']<=300
condition2_df = condition1_df[df_mask]

condition2_df
```

### 4. Output the list of movies in decreasing order of their scores.

Ici, nous trions directement la base de données sur les critères que nous avons selectionnées de manière à réduire les calculs par la suite.  

   - Il nous reste maintenant à appliquer notre metric de la plus haute note : on trie notre base en dans l'ordre décroissant par rapport à la note moyenne du film
    

```python
#Affichage du Dataframe avec la metric de la meilleure note
score_df = condition2_df.sort_values(by=['vote_average'], ascending = False)

score_df
```

On observe que dans notre DataFrame, certaines notes sont inférieur à la moyenne, dans le cadre d'un système de recommandation, nous souhaitons recommmander des films dont la note moyenne est supérieur à 6, cela réduira aussi le DataFrame.

```python
df_mask=film_df['vote_average']>=6
condition3_df = score_df[df_mask]

condition3_df
```

```python
score1_df = condition3_df[['title','vote_average']]

print ("Scores triés dans l'ordre décroissant : ")
print(score1_df)
```

Nous retrouvons donc bien nos films avec une note moyenne supérieure à 6, avec un nombre de vote supérieur ou égal à 50, avec une durée comprise entre 45 et 300min.  

### Les premiers films affichés sont ceux qui ont la note moyenne la plus haute qui représente aussi le score que nous attribuons au film.

# Calculating Score and output

### 1. The last value to calculate is C which is the mean rating for all the movies in the dataset.

```python
#Calcul de la moyenne du vote average dans notre cas
C = score1_df['vote_average'].mean()
C
```

Ici, la moyenne est assez haute car nous avons trié notre Dataframe par moyennes générale supérieures à 6. 

  - Afin de calculer le score en fonction de la moyenne, nous allons simplement soustraire C à chaque note du Dataframe, et supprimer les valeurs négatives.

### 2. Calculate the score.

```python
#Création de base de données avec la nouvelle colonnes des scores

score2_df=score1_df.assign(score=0)

score2_df
```

### 3. Add the score as a feature to the dataframe, by using Apply function.

```python
def score_c(x):
    C = 6.82
    return x - C
```

```python
#On remplis notre colonne avec la fonction apply

score2_df['score'] = score2_df['vote_average'].apply(score_c)
score2_df
```

### 4. Sort the dataframe on the basis of the score just computed.

Après avoir effectué le calcul du score, on supprime les films qui ont des scores négatifs

```python
#Suppression des films négatifs

df_mask=score2_df['score']>0
condition4_df = score2_df[df_mask]

condition4_df
```

```python
score3_df = condition4_df[['title','score']]

print ("Scores triés dans l'ordre décroissant : ")
print(score3_df)
```

### 5. Output the list of top movies

```python
#On tri notre DataFrame en fonction du score
score3_df = condition4_df.sort_values(by=['score'], ascending = False)

score3_df
```

Il n'y à pas de changement au sein du Dataframe car nos données sont des soustractions du vote_average sur un Dataframe déjà trié dans l'ordre décroissant par rapport au vote_average.

### 2. This matrix will be sparse (lot of unknown values), explain why. Propose a way to fill in the missing values and justify it. Implement it to have a matrix without missing values.

```python
pivoted = matrix.pivot(index="userId", columns="movieId", values="rating")
```

```python
pivoted
```

La somme totale de valeurs manquantes ici est expliqué du fait que nous avons 671 personnes pour 9066 films, il est probable que les utilisateurs n'ont pas vus tous les films. 
   - Nous proposons de remplacer nos valeurs manquantes par une imputation par les plus proches voisins afin d'attribuer une note à une valeur manquante associée à l'échantillon par une valeur qui lui ressemble le plus dans l'échantillon. Ici, la note manquante d'un utilisateur sera remplacé par une note égale à un autre utilisateur qui à les mêmes valeurs.

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors = 1)

matrix = imputer.fit_transform(pivoted)

```

```python
matrix = pd.DataFrame(matrix)

matrix
```

```python
matrix.isnull().sum()
```

### 4. Implement a function to return the n nearest neighbors of a given user in the sense of the calculated similarity.

```python
#Fonction permettant de renvoyer les n voisins les plus proches d'un utilisateur selectionné
def find_n_neighbours(df,n):
    order = np.argsort(df.values, axis=1)[:, :n]
    df = df.apply(lambda x: pd.Series(x.sort_values(ascending=False)
           .iloc[:n].index, 
          index=['top{}'.format(i) for i in range(1, n+1)]), axis=1)
    return df
```

```python
find_n_neighbours(pivoted1.iloc[129:130,:], 3)
```

Ici on renvoie les user qui sont suceptible d'avoir les mêmes goûts que l'utilisateur 5
