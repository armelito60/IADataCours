## Part 2 : Building a simple recommender

Les pré-requis : 

La valeur v -> vote count

La valeur R -> vote average

La Valeur m est un choix arbitraire

```python
df  = pd.read_csv('movies_metadata.csv')
```

```python
df
```

```python
m = df['vote_count'].quantile(0.8)
m
```

```python
C = df['vote_average'].mean()
C
```

```python
#Tous les films qui satisfaient nos conditions
movie_conditions = df.copy().loc[df['vote_count'] >= m]
```

```python
#Calculate the weighted rating score
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    return (v/(v+m) * R) + (m/(m+v) * C)

#Add the score as a feature to the dataframe
df['score'] = df.apply(weighted_rating, axis = 1)
```

```python
#tri du dataframe et affichage des top films en fonction du score
df.groupby(["title"])['score'].max().sort_values(ascending=False)[:10]
```

## Part 3 : Implement a user-based collaborative filtering system

```python
ratings_df = pd.read_csv('ratings_small.csv')

```

```python
ratings_df.head()
```

```python
pivoted = ratings_df.pivot(index='userId', columns='movieId', values='rating')
```

```python
pivoted.head()
```

Beaucoup de valeurs non connues car les users n'ont soit pas vu le film ou soit n'ont pas donné d'avis au film visionné.

```python
pivoted = pivoted.fillna(0)
pivoted.head()
```

#### On enlève le bruit

- On ne veut pas des films qui sont notés par un petit nombre de user.
- De même, les users qui ont notés peut de films ne seront pas pris en compte

On prend en considération :
- pour qualifier un film : minimum 10 users doivent avoir voté un film
- pour qualifier un user : minimum 50 films doivent avoir été vus par le user

```python
no_user_voted = ratings_df.groupby('movieId')['rating'].agg('count')
no_user_voted
```

```python
no_movies_voted = ratings_df.groupby('userId')['rating'].agg('count')
no_movies_voted
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# user similarity on replacing NaN by 0.0
b = cosine_similarity(pivoted)
np.fill_diagonal(b, 0)
similarity_with_user = pd.DataFrame(b,index=pivoted.index)
similarity_with_user.columns=pivoted.index
similarity_with_user.head()
```

```python
def find_n_neighbours(df,n):
    order = np.argsort(df.values, axis=1)[:, :n]
    df = df.apply(lambda x: pd.Series(x.sort_values(ascending=False)
           .iloc[:n].index, 
          index=['top{}'.format(i) for i in range(1, n+1)]), axis=1)
    return df

# display top 30 neighbors for the first user
find_n_neighbours(similarity_with_user.iloc[:1,:], 10)
```
