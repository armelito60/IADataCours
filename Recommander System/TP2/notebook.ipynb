## Part 2 : Calculating centrality metrics using Python

```python
G1 = nwx.Graph([(7,2), (2,3), (7,4), (4,5), (7,3), (7,5), (1,6),(1,7),(2,8),(2,9)])
nwx.draw(G1, with_labels = True)
plt.show()
```

```python
deg_centrality = nwx.degree_centrality(G1)
deg_centrality
```

C'est le nombre de lien direct divisé par (n-1), n correspondant au nombre de noeuds.

```python
bet_centrality = nwx.betweenness_centrality(G1)
bet_centrality
```

Plus la valeur du noeud est proche de 1, plus le nombre du plus court chemin est élevé.

```python
close_centrality = nwx.closeness_centrality(G1)
close_centrality
```

C'est (n-1) où n est le nombre total de noeud divisé par le nombre d'arrêtes reliant un point donné à tous les autres.
Donc plus la valeur est proche de 1, plus le noeud est proche de tous les autres.
