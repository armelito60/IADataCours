## Part 2 : Building a simple recommender

Les pré-requis : 

La valeur v -> vote count

La valeur R -> vote average

La Valeur m est un choix arbitraire

```python
df  = pd.read_csv('movies_metadata.csv')
```

```python
df
```

```python
m = df['vote_count'].quantile(0.8)
m
```

```python
C = df['vote_average'].mean()
C
```

```python
#Tous les films qui satisfaient nos conditions
movie_conditions = df.copy().loc[df['vote_count'] >= m]
```

```python
#Calculate the weighted rating score
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    return (v/(v+m) * R) + (m/(m+v) * C)

#Add the score as a feature to the dataframe
df['score'] = df.apply(weighted_rating, axis = 1)
```

```python
#tri du dataframe et affichage des top films en fonction du score
df.groupby(["title"])['score'].max().sort_values(ascending=False)[:10]
```
